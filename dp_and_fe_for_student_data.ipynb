```python
print("\nüìä ‡∏™‡∏£‡∏∏‡∏õ‡∏õ‡∏±‡∏ç‡∏´‡∏≤‡∏ó‡∏µ‡πà‡∏û‡∏ö‡πÉ‡∏ô‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•:")
print(f"- ‡∏Ñ‡πà‡∏≤‡∏´‡∏≤‡∏¢‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î: {df.isnull().sum().sum()} ‡∏Ñ‡πà‡∏≤")
print(f"- ‡∏≠‡∏≤‡∏¢‡∏∏‡∏ú‡∏¥‡∏î‡∏õ‡∏Å‡∏ï‡∏¥: {df[df['‡∏≠‡∏≤‡∏¢‡∏∏'] > 100]['‡∏≠‡∏≤‡∏¢‡∏∏'].values if any(df['‡∏≠‡∏≤‡∏¢‡∏∏'] > 100) else '‡πÑ‡∏°‡πà‡∏°‡∏µ'}")
print(f"- GPA ‡∏ú‡∏¥‡∏î‡∏õ‡∏Å‡∏ï‡∏¥: {df[(df['GPA'] < 0) | (df['GPA'] > 4)]['GPA'].values if any((df['GPA'] < 0) | (df['GPA'] > 4)) else '‡πÑ‡∏°‡πà‡∏°‡∏µ'}")
print(f"- ‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö‡∏Å‡∏≤‡∏£‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡∏Ñ‡∏ì‡∏∞: {df['‡∏Ñ‡∏ì‡∏∞'].nunique()} ‡πÅ‡∏ö‡∏ö")
print(f"- ‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö‡∏Å‡∏≤‡∏£‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡πÄ‡∏û‡∏®: {df['‡πÄ‡∏û‡∏®'].nunique()} ‡πÅ‡∏ö‡∏ö")
```

‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡∏ú‡∏¥‡∏î‡∏õ‡∏Å‡∏ï‡∏¥‡πÉ‡∏ô filed ‡∏Ç‡∏≠‡∏á‡∏≠‡∏≤‡∏¢‡∏∏‡πÇ‡∏î‡∏¢‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ ‡πÉ‡∏ä‡πâ IQR (Interquartile Range)

```python
print(f"\n‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏≠‡∏≤‡∏¢‡∏∏ (‡∏£‡∏ß‡∏°‡∏Ñ‡πà‡∏≤‡∏ú‡∏¥‡∏î‡∏õ‡∏Å‡∏ï‡∏¥) - {df['‡∏≠‡∏≤‡∏¢‡∏∏'].mean():.2f}")
Q1 = df['‡∏≠‡∏≤‡∏¢‡∏∏'].quantile(0.25)
Q3 = df['‡∏≠‡∏≤‡∏¢‡∏∏'].quantile(0.75)
IQR = Q3 - Q1
```

‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏Ç‡∏≠‡∏ö‡πÄ‡∏Ç‡∏ï

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[(df['‡∏≠‡∏≤‡∏¢‡∏∏'] >= lower_bound) & (df['‡∏≠‡∏≤‡∏¢‡∏∏'] <= upper_bound)]
```

‡∏Å‡∏£‡∏≠‡∏á‡∏Ñ‡πà‡∏≤‡∏ú‡∏¥‡∏î‡∏õ‡∏Å‡∏ï‡∏¥‡∏Ç‡∏≠‡∏á‡∏≠‡∏≤‡∏¢‡∏∏

```python
print(f"\n‡∏Ç‡∏≠‡∏ö‡πÄ‡∏Ç‡∏ï‡∏•‡πà‡∏≤‡∏á - {lower_bound}")
print(f"‡∏Ç‡∏≠‡∏ö‡πÄ‡∏Ç‡∏ï‡∏ö‡∏ô - {upper_bound}")
print(f"\n‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏´‡∏•‡∏±‡∏á‡∏Å‡∏£‡∏≠‡∏á‡∏Ñ‡πà‡∏≤‡∏ú‡∏¥‡∏î‡∏õ‡∏Å‡∏ï‡∏¥")
print(df)
print(f"\n‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏≠‡∏≤‡∏¢‡∏∏ (‡∏´‡∏•‡∏±‡∏á‡∏Å‡∏£‡∏≠‡∏á‡∏Ñ‡πà‡∏≤‡∏ú‡∏¥‡∏î‡∏õ‡∏Å‡∏ï‡∏¥) - {df['‡∏≠‡∏≤‡∏¢‡∏∏'].mean():.2f}")
```

‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÄ‡∏û‡∏®‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏Å‡∏≤‡∏£‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡∏´‡∏•‡∏≤‡∏¢‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö‡πÉ‡∏´‡πâ‡∏°‡∏µ‡∏°‡∏≤‡∏ï‡∏£‡∏≤‡∏ê‡∏≤‡∏ô‡πÄ‡∏î‡∏µ‡∏¢‡∏ß‡∏Å‡∏±‡∏ô ‡πÅ‡∏•‡∏∞‡∏ñ‡πâ‡∏≤‡∏´‡∏≤‡∏Å‡πÄ‡∏õ‡πá‡∏ô‡∏Ñ‡πà‡∏≤ NaN ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô ‡πÑ‡∏°‡πà‡∏£‡∏∞‡∏ö‡∏∏

```python
def standardize_gender(name):
    if name in ['‡∏ä‡∏≤‡∏¢', '‡∏ä', 'M', 'Male']:
        return '‡∏ä‡∏≤‡∏¢'
    elif name in ['‡∏´‡∏ç‡∏¥‡∏á', 'Female', '‡∏ç', 'F']:
        return '‡∏´‡∏ç‡∏¥‡∏á'
    else:
        return '‡πÑ‡∏°‡πà‡∏£‡∏∞‡∏ö‡∏∏'
```

```python
df['‡πÄ‡∏û‡∏®'] = df['‡πÄ‡∏û‡∏®'].apply(standardize_gender)
df
```

‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏Ñ‡∏ì‡∏∞‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏Å‡∏≤‡∏£‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡∏´‡∏•‡∏≤‡∏¢‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö‡πÉ‡∏´‡πâ‡∏°‡∏µ‡∏°‡∏≤‡∏ï‡∏£‡∏≤‡∏ê‡∏≤‡∏ô‡πÄ‡∏î‡∏µ‡∏¢‡∏ß‡∏Å‡∏±‡∏ô ‡πÅ‡∏•‡∏∞‡∏ñ‡πâ‡∏≤‡∏´‡∏≤‡∏Å‡πÄ‡∏õ‡πá‡∏ô‡∏Ñ‡πà‡∏≤ NaN ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô ‡πÑ‡∏°‡πà‡∏£‡∏∞‡∏ö‡∏∏

```python
def standardize_faculty(name):
    if name in ['‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå', '‡∏ß‡∏¥‡∏ó‡∏¢‡πå', 'Science', '‡∏Ñ‡∏ì‡∏∞‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå', 'Sci']:
        return '‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå'
    elif name in ['‡∏®‡∏¥‡∏•‡∏õ‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå', '‡∏®‡∏¥‡∏•‡∏õ‡πå', 'Arts', 'Liberal Arts']:
        return '‡∏®‡∏¥‡∏•‡∏õ‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå'
    elif name in ['‡∏ß‡∏¥‡∏®‡∏ß‡∏Å‡∏£‡∏£‡∏°‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå', '‡∏ß‡∏¥‡∏®‡∏ß‡∏∞', '‡∏ß‡∏¥‡∏®‡∏ß‡∏Ø', 'Engineering', '‡∏ß‡∏¥‡∏®‡∏ß‡∏Å‡∏£‡∏£‡∏°']:
        return '‡∏ß‡∏¥‡∏®‡∏ß‡∏Å‡∏£‡∏£‡∏°‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå'
    else:
        return '‡πÑ‡∏°‡πà‡∏£‡∏∞‡∏ö‡∏∏'
```

```python
df['‡∏Ñ‡∏ì‡∏∞'] = df['‡∏Ñ‡∏ì‡∏∞'].apply(standardize_faculty)
df
```

‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç‡∏Ñ‡πà‡∏≤ GPA ‡∏ó‡∏µ‡πà‡∏ú‡∏¥‡∏î‡∏õ‡∏Å‡∏ï‡∏¥(‡∏°‡∏µ‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡πá‡∏ô NaN, <1, >4)
‡πÇ‡∏î‡∏¢ ‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà GPA ‡∏ï‡∏¥‡∏î‡∏•‡∏ö ‡∏à‡∏∞‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô‡∏Ñ‡πà‡∏≤‡πÄ‡∏õ‡πá‡∏ô NaN
‡∏™‡πà‡∏ß‡∏ô‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà GPA ‡πÄ‡∏Å‡∏¥‡∏ô 4.0 ‡∏à‡∏∞‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô‡∏Ñ‡πà‡∏≤‡πÄ‡∏õ‡πá‡∏ô 4.0 ‡πÅ‡∏•‡∏∞‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢ GPA ‡∏à‡∏≤‡∏Å‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á ‡πÅ‡∏•‡πâ‡∏ß‡πÄ‡∏ï‡∏¥‡∏°‡πÉ‡∏ô‡∏ä‡πà‡∏≠‡∏á‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡πá‡∏ô NaN

```python
df.loc[df['GPA'] < 0, 'GPA'] = np.nan
df.loc[df['GPA'] > 4, 'GPA'] = 4.0
mean_gpa = df['GPA'].mean().round(2)
df['GPA'].fillna(mean_gpa, inplace=True)

df
```

‡∏õ‡∏£‡∏∞‡∏°‡∏≤‡∏ô‡∏Ñ‡πà‡∏≤‡∏Ç‡∏≠‡∏á‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å‡πÇ‡∏î‡∏¢‡πÉ‡∏ä‡πâ‡∏à‡∏≤‡∏Å‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏à‡∏≤‡∏Å‡πÉ‡∏ô‡∏ï‡∏≤‡∏£‡∏≤‡∏á

```python
mean_weight = df['‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å_kg'].mean().round(2)
df['‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å_kg'].fillna(mean_weight, inplace=True)
print(f"\n‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å - {mean_weight:.2f} ‡∏Å‡∏Å.")
```

#Feature Engineering - ‡∏Å‡∏≤‡∏£‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ñ‡∏∏‡∏ì‡∏•‡∏±‡∏Å‡∏©‡∏ì‡∏∞‡πÉ‡∏´‡∏°‡πà

‡∏™‡∏£‡πâ‡∏≤‡∏á Feature ‡∏à‡∏≤‡∏Å‡∏™‡πà‡∏ß‡∏ô‡∏™‡∏π‡∏á ‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å ‡∏´‡∏≤‡∏Ñ‡πà‡∏≤ BMI

```python
df['BMI'] = df['‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å_kg'] / (df['‡∏™‡πà‡∏ß‡∏ô‡∏™‡∏π‡∏á_cm']/100)**2
df['‡∏™‡∏∏‡∏Ç‡∏†‡∏≤‡∏û'] = pd.cut(df['BMI'],
                      bins=[0, 18.5, 25, 30, 100],
                      labels=['‡∏ú‡∏≠‡∏°', '‡∏õ‡∏Å‡∏ï‡∏¥', '‡∏≠‡πâ‡∏ß‡∏ô', '‡∏≠‡πâ‡∏ß‡∏ô‡∏°‡∏≤‡∏Å'])

print(df[['‡∏ä‡∏∑‡πà‡∏≠', 'BMI', '‡∏™‡∏∏‡∏Ç‡∏†‡∏≤‡∏û']])
```
